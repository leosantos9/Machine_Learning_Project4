```python
import pandas as pd
```

```python
a_df = pd.read_csv("../Data/answers.csv")
c_df = pd.read_csv("../Data/comment_score.csv")
a_df
```

```python
df = pd.merge(a_df, c_df, on='answers_id', how="left")
df
```

```python
new_df = df.filter(['answers_question_id', 'answers_date_added','answers_id','answers_author_id','score','comments'])
new_df["score"].fillna(0, inplace = True) 
new_df["comments"].fillna(0, inplace = True) 

new_df
```

```python
df = new_df.sort_values(by = ['answers_question_id', 'answers_date_added'], ascending = [False, True])

df.head
```

```python
df['q_score'] = (df.answers_question_id != df.answers_question_id.shift()).astype(int)
df.loc[0,'q_score'] = 0
```

```python
df
```

```python
df['n_score'] = df['score']
df.loc[df['score'] > 0, 'n_score'] = 1

df
```

```python
df['total_score'] = df.apply(lambda row: row['comments' : 'n_score'].sum(),axis=1)
```

```python
df.head()
```

```python
df_f = df.sort_values(by = ['answers_question_id','total_score'], ascending = [False, True])
```

```python
df_f
```

```python
final_db = df.loc[df.groupby('answers_question_id')['total_score'].idxmax().values.ravel()]
```

```python
final_db
```

```python
p_df = pd.read_csv("../Data/professionals.csv")
```

```python
p_df.head()
```

```python
p_df.rename(columns={'professionals_id':'answers_author_id'}, 
                 inplace=True)
```

```python
p_df["professionals_industry"].fillna("Other", inplace = True) 
```

```python
p_df.head()
```

```python
df = pd.merge(p_df, final_db, on='answers_author_id')
df.head
```

```python
clean_df = df.filter(['answers_question_id','answers_id','answers_question_id','professionals_industry'])

prof = clean_df.sort_values(by = ['professionals_industry'], ascending = True)
```

```python
p = prof.groupby('professionals_industry').count()
p['new_col'] = range(1, len(p) + 1)

p.head(50)
```

```python
pp = p.drop(['answers_question_id','answers_id'], axis = 1)
```

```python
pp
```

```python
pp.to_csv('../Data/professionals_index_new.csv', header= False)
```

```python
pros_dict = pp.to_dict().get("new_col")
industry_codes = []

for index, row in clean_df.iterrows():
    industry_codes.append(pros_dict[row['professionals_industry']])
    
clean_df["industry_codes"] = industry_codes
clean_df = clean_df.rename(index=str, columns={"answers_question_id": "questions_id"})

clean_df.to_csv('../Data/data_with_industry_codes.csv', header= True)
```
